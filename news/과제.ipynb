# 특성추출을 바탕으로 분석

```python
# import warnings
# warnings.filterwarnings(action='ignore') 
# import pandas as pd
# import numpy as np
# df = pd.read_csv('전체 뉴스.csv') 
df['date'] = pd.to_datetime(df['일자'].astype(str), format='%Y%m%d')

df_qy8=df.query("date >= '2021-04-1' and date <='2021-07-26'")

new_data=df_qy8['특성추출(가중치순 상위 50개)'].values.tolist()
keyword=[]
for i in new_data:
    for y in i.split(','):
        keyword.append(y)
#단어별 빈도수 계산
from collections import Counter
count=Counter(keyword)
tags = count.most_common(30)
print(dict(tags))

# 카운터 파일 저장 - 삭제 가능 (안해도 무관)
aa=dict(tags)
with open('2021카운터.csv', 'w') as f:
    for key in aa.keys():
        f.write("%s, %s\n" % (key, aa[key]))
    


#워드클라우드를 이용한 단어 빈도수 시각화
# WordCloud 설치가 안된 경우, pip install wordcloud
import matplotlib.pyplot as plt
from wordcloud import WordCloud
wc = WordCloud(font_path='malgun.ttf',background_color='white', width=1000, height=800)
cloud = wc.generate_from_frequencies(aa)
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
```

```python
new_data=df_qy19['특성추출(가중치순 상위 50개)'].values.tolist()
keyword=[]
nouns=''
for i in new_data:
    for y in i.split(','):
        nouns +=y+' '
    keyword.append(nouns)
    nouns=''

#TF-IDF 계산

from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF vectorizer
tfv = TfidfVectorizer(ngram_range = (1,1))
#Transformation
vec_text = tfv.fit_transform(keyword)
#Features
words = tfv.get_feature_names()
#KMeans 계산
from sklearn.cluster import KMeans
## K Means Clustering
kmeans = KMeans(n_clusters = 2, n_init = 17, n_jobs = -1, tol = 0.01, max_iter = 200)
#Fitting
kmeans.fit(vec_text)
## Transforming Numbers Back To Words
common_words = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```

# 키워드를 바탕으로 분석

```python
# import warnings
# warnings.filterwarnings(action='ignore') 

# import pandas as pd
# import numpy as np

# df = pd.read_csv('전체 뉴스.csv') 
df['date'] = pd.to_datetime(df['일자'].astype(str), format='%Y%m%d')

#df_qy8=df.query("date >= '2008-01-1' and date <='2010-12-31'")

df_qy19=df.query("date >= '2021-04-21' and date <='2021-07-26'")

new_data=df_qy19['키워드'].values.tolist()
keyword=[]
for i in new_data:
    for y in i.split(','):
        keyword.append(y)
    
#단어별 빈도수 계산
from collections import Counter
count=Counter(keyword)
tags = count.most_common(30)
print(tags)
bb=dict(tags)

#워드클라우드를 이용한 단어 빈도수 시각화
import matplotlib.pyplot as plt
from wordcloud import WordCloud
wc = WordCloud(font_path='malgun.ttf',background_color='white', width=1000, height=800)
cloud = wc.generate_from_frequencies(dict(tags))
plt.figure(figsize=(10,8))
plt.axis('off')
plt.imshow(cloud)
```

```python
new_data=df_qy19['키워드'].values.tolist()
keyword=[]
nouns=''
for i in new_data:
    for y in i.split(','):
        nouns +=y+' '
    keyword.append(nouns)
    nouns=''

#TF-IDF 계산

from sklearn.feature_extraction.text import TfidfVectorizer
#TF-IDF vectorizer
tfv = TfidfVectorizer(ngram_range = (1,1))
#Transformation
vec_text = tfv.fit_transform(keyword)
#Features
words = tfv.get_feature_names()
#KMeans 계산
from sklearn.cluster import KMeans
## K Means Clustering
kmeans = KMeans(n_clusters = 2, n_init = 17, n_jobs = -1, tol = 0.01, max_iter = 200)
#Fitting
kmeans.fit(vec_text)
## Transforming Numbers Back To Words
common_words = kmeans.cluster_centers_.argsort()[:,-1:-11:-1]
for num, centroid in enumerate(common_words):
    print(str(num) + ' : ' + ', '.join(words[word] for word in centroid))
```
